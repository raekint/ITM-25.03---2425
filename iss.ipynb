```python
import requests
import matplotlib.pyplot as plt
import time

latitudes = []
longitudes = []

for data_point in range(100):
    ping = requests.get("https://api.wheretheiss.at/v1/satellites/25544")
    data = ping.json()

    latitudes.append(data['latitude'])
    longitudes.append(data['longitude'])

    time.sleep(10)

plt.figure(figsize=(10, 10))
plt.plot(longitudes, latitudes, marker='o', linestyle='-', color='r')

plt.title('Track the ISS')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```
